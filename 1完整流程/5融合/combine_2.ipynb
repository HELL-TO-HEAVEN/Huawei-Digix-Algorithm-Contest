```python
import pandas as pd
```

# lgb

```python
# 三天50ｗ融合
lgb_result = pd.read_csv('lgb-0.964666.csv')
sub1 = pd.DataFrame()
sub1['probability'] = lgb_result['probability']
```

```python
# 单天最高
lgb_result = pd.read_csv('75w-0.964869.csv')
sub1 = pd.DataFrame()
sub1['probability'] = lgb_result['probability']
```

# xgb

```python
xgb_result = pd.read_csv('xgb-0.964106.csv')
sub2 = pd.DataFrame()
sub2['probability'] = xgb_result['probability']
```

# xDeepFM

```python
xDeepFM_result = pd.read_csv('xdeepfm_0.958468.csv')
sub3 = pd.DataFrame()
sub3['probability'] = xDeepFM_result['probability']
```

# 融合(权重需要试)

# 0.4-0.4-0.2   score:0.963799

```python
sub = pd.DataFrame({'id':lgb_result['id']})
sub['probability'] = (0.4*sub1['probability'] + 
                      0.4*sub2['probability'] + 
                      0.2*sub3['probability']).round(6)

sub.to_csv('submission_0.4_0.4_0.2.csv',index = False)
```

# 加权平均   score:0.963176

```python
sub = pd.DataFrame({'id':lgb_result['id']})
sub['probability'] = (sub1['probability'] + 
                      sub2['probability'] + 
                      sub3['probability'])/3
sub['probability'] = sub['probability'].round(6)

sub.to_csv('submission_0.33_0.33_0.33.csv',index = False)
```

# 0.964886 和其他

```python
sub = pd.DataFrame({'id':lgb_result['id']})
sub['probability'] = (0.6*sub1['probability'] + 
                      0.3*sub2['probability'] + 
                      0.1*sub3['probability']).round(6)

sub.to_csv('submission_0.6_0.3_0.1.csv',index = False)
```

# 投票

```python
sub = pd.DataFrame({'id':lgb_result['id']})
sub['probability1'] = lgb_result['probability']
sub['probability2'] = xgb_result['probability']
sub['probability3'] = xDeepFM_result['probability']
sub.head(20)
```

```python
cols = ['probability1','probability2','probability3']
```

```python
sub[sub[cols] < 0.5] = 0
sub[sub[cols] >= 0.5] = 1
```

```python
sub.head(20)
```

```python
sub['count'] = sub[cols].sum(axis = 1)
```

```python
sub.head(20)
```

```python
sub.loc[sub['count'] >= 2,'label'] = 1
sub.loc[sub['count'] < 2,'label'] = 0
```

```python
sub.head(20)
```

```python
sub['probability1'] = lgb_result['probability']
sub['probability2'] = xgb_result['probability']
sub['probability3'] = xDeepFM_result['probability']
```

```python
sub.head(20)
```

```python
sub.loc[sub['label'] == 1,'probability'] = sub[cols].max(axis = 1)
sub.loc[sub['label']== 0,'probability'] = sub[cols].min(axis = 1)
```

```python
sub.head(50)
```

```python
submission = pd.DataFrame({'id':sub['id'],'probability':sub['probability']})
```

```python
submission['probability']=submission['probability'].round(6)
```

```python
submission.to_csv('submission_vote_2.csv',index = False)
```
